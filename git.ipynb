```python
""" Знакомство с командами git. """
```

1. **git --version** — проверка текущей версии Git.
2. **git init** — инициализация нового локального репозитория.
3. **git clone <url>** — клонирование удаленного репозитория на локальную машину.
4. **git add** — добавление файлов в индекс:
   - `git add <file>` — добавление конкретного файла.
   - `git add *.py` — добавление всех файлов с расширением `.py`.
   - `git add .` — добавление всех файлов в текущей директории.
5. **git commit -m 'сообщение'** — создание коммита с сообщением.
6. **git commit -am 'сообщение'** — добавление изменений и создание коммита (без отдельного `git add`).
7. **git commit --amend -m 'новое_сообщение'** — изменение сообщения последнего коммита.
8. **git status** — просмотр текущего состояния репозитория.
9. **git config** — настройка пользователя:
   - `git config --global user.name 'Ваше_имя'` — установка имени.
   - `git config --global user.email 'ваша_почта@example.com'` — установка почты.
10. **git diff** — просмотр изменений:
    - `git diff` — сравнение рабочей директории и индекса.
    - `git diff <file>` — изменения в конкретном файле.
    - `git difftool` — просмотр изменений в графическом интерфейсе.
11. **git log** — просмотр истории коммитов:
    - `git log --oneline` — краткий вывод коммитов.
    - `git log --oneline -n 3` — последние 3 коммита.
    - `git show <хэш>:<файл>` — содержимое файла на момент коммита.
12. **git rm <file>** — удаление файла из репозитория и индекса.
13. **git mv <старый_файл> <новый_файл>** — переименование или перемещение файла.
14. **git reset** — отмена изменений:
    - `git reset HEAD <file>` — отмена индексации файла.
15. **git restore** — отмена несохраненных изменений:
    - `git restore <file>` — откат файла к последнему коммиту.
    - `git restore --staged <file>` — удаление файла из индекса.
16. **git revert -n HEAD~3..HEAD** — отмена последних 3 коммитов.
17. **git checkout <хэш> -- <file>** — возврат файла к состоянию конкретного коммита.
18. **git remote add origin <url>** — связывание локального репозитория с удаленным.
19. **git push -u origin master** — первая отправка изменений на удаленный репозиторий.
20. **git push** — отправка изменений на удаленный репозиторий.
21. **git help <command>** — справка по конкретной команде.
22. **git ls-files --other --ignored --exclude-standard** — список игнорируемых файлов.
